## Ingesting data into Bronze Lakehouse tables

```python
#Lakehouse files path ingested from data pipeline (parameter)
files_path = "samplepath"
```

```python
#Reading the CSV file (in shortcut file) into the dataframes
df_students = spark.read.option("header", True).csv(f"{files_path}/students.csv")
df_assignments = spark.read.option("header", True).csv(f"{files_path}/assignments.csv")
df_courses = spark.read.option("header", True).csv(f"{files_path}/courses.csv")
df_submissions = spark.read.option("header", True).csv(f"{files_path}/submissions.csv")
df_enrollments = spark.read.option("header", True).csv(f"{files_path}/enrollments.csv")

#also tells spark that the first row contains column headers
```

```python
#Writing the CSV files to bronze lakehouse tables
df_students.write.mode("overwrite").format("delta").saveAsTable("bronze_students")
df_assignments.write.mode("overwrite").format("delta").saveAsTable("bronze_assignments")
df_courses.write.mode("overwrite").format("delta").saveAsTable("bronze_courses")
df_submissions.write.mode("overwrite").format("delta").saveAsTable("bronze_submissions")
df_enrollments.write.mode("overwrite").format("delta").saveAsTable("bronze_enrollments")
```
